```python
import pandas as pd
import os

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

from selenium.webdriver.common.by import By # By es para buscar por tag, clase, id...
from selenium.webdriver.common.keys import Keys  

import time
from tqdm.notebook import tqdm

import warnings
warnings.filterwarnings('ignore')
```

```python
#driver configuration
opciones=Options()

opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.headless=False    # si True, no aperece la ventana (headless=no visible)
opciones.add_argument('--start-maximized')         # comienza maximizado
#opciones.add_argument('user-data-dir=selenium')    # mantiene las cookies
#opciones.add_extension('driver_folder/adblock.crx')       # adblocker
opciones.add_argument('--incognito')
```

```python
#driver = "./chromedriver.exe" #remember substitute this for your driver path
#driver = webdriver.Chrome(driver,options = opciones)
PATH = ChromeDriverManager().install()     # instala el driver de chrome

driver=webdriver.Chrome(PATH, options = opciones)  
```

```python
url = 'https://www.nflpenalties.com/week/1?year=2022&view=log'

driver.get(url)
```

```python
time.sleep(3)

driver.find_element(By.XPATH, '/html/body/div[5]/div[2]/div[1]/div[2]/div[2]/button[2]').click()
```

```python
len(driver.find_elements(By.XPATH, '//th[@class="footable-sortable"]'))
```

```python
cols = [e.text for e in driver.find_elements(By.XPATH, '//th[@class="footable-sortable"]')]
len(cols)
```

```python
celdas = driver.find_elements(By.XPATH, '//td')[0:-16]
len(celdas)
```

```python
penalty = [celdas[i].text for i in range(0, len(celdas), 16)]
date = [celdas[i].text for i in range(1, len(celdas), 16)]
against = [celdas[i].text for i in range(2, len(celdas), 16)]
beneficiary = [celdas[i].text for i in range(3, len(celdas), 16)]
player = [celdas[i].text for i in range(4, len(celdas), 16)]
pos = [celdas[i].text for i in range(5, len(celdas), 16)]
ref_crew = [celdas[i].text for i in range(6, len(celdas), 16)]
quarter = [celdas[i].text for i in range(7, len(celdas), 16)]
time = [celdas[i].text for i in range(8, len(celdas), 16)]
down = [celdas[i].text for i in range(9, len(celdas), 16)]
distance = [celdas[i].text for i in range(10, len(celdas), 16)]
declined = [celdas[i].text for i in range(11, len(celdas), 16)]
offsetting = [celdas[i].text for i in range(12, len(celdas), 16)]
yards = [celdas[i].text for i in range(13, len(celdas), 16)]
home = [celdas[i].text for i in range(14, len(celdas), 16)]
phase = [celdas[i].text for i in tqdm(range(15, len(celdas), 16))]
```

```python
df = pd.DataFrame(columns=cols)
```

```python
df['Penalty'] = penalty
df['Date'] = date
df['Against'] = against
df['Beneficiary'] = beneficiary
df['Player'] = player
df['Pos'] = pos
df['Ref Crew'] = ref_crew
df['Quarter'] = quarter
df['Time'] = time
df['Down'] = down
df['Distance'] = distance
df['Declined'] = declined
df['Offsetting'] = offsetting
df['Yards'] = yards
df['Home'] = home
df['Phase'] = phase
```

```python
winners = [e for e in df3['Winner/tie']]
losers = [e for e in df3['Loser/tie']]
```

```python
df['Winner'] = None
df['Loser'] = None

for i in range(len(df['Against'])):
    if df['Against'][i] in winners:
        df['Winner'][i] = df['Against'][i]
    elif df['Against'][i] in losers:
        df['Loser'][i] = df['Against'][i]

for i in range(len(df['Beneficiary'])):
    if df['Beneficiary'][i] in winners:
        df['Winner'][i] = df['Beneficiary'][i]
    elif df['Beneficiary'][i] in losers:
        df['Loser'][i] = df['Beneficiary'][i]
```

```python
df4 = pd.read_csv('../2022 NFL Schedule - Week 1.csv')
```

```python
df4 = df4.rename(columns={
    "Unnamed: 0": "cero",
    "Unnamed: 1": "Away Team",
    "Unnamed: 2": "dos",
    "Unnamed: 3": "Home Team",
    "Unnamed: 4": "cuatro",
    "Unnamed: 5": "Time"
})
```

```python
df4 = df4.dropna(axis=1, how='all')
df4 = df4.drop_duplicates()
df4 = df4.drop(columns = ["cero", "dos", "cuatro"])
df4 = df4.dropna(axis=0, how='all')
df4 = df4.reset_index(drop=True)
```

```python
df4['Primetime'] = 'NO'

for i in [1, 16, 18]:
    df4.loc[i, 'Primetime'] = 'YES'
```

```python
df4.drop([0, 2, 17], inplace=True)
```

```python
df4.reset_index(drop=True, inplace=True)
```

```python
df4.head()
```

```python
df4['Away Team'] = df4['Away Team'].map(team_map_df3)
df4['Home Team'] = df4['Home Team'].map(team_map_df3)
```

```python
mask1 = df4['Primetime'] == 'YES'
primetimers1 = df4.loc[mask1, 'Home Team'].tolist()

mask2 = df4['Primetime'] == 'YES'
primetimers2 = df4.loc[mask2, 'Away Team'].tolist()

primetimers = primetimers1 + primetimers2
primetimers
```

```python
df['Prime Time'] = None

for i in range(len(df['Prime Time'])):
    if df['Beneficiary'][i] in primetimers:
        df['Prime Time'][i] = 'YES'
    else:
        df['Prime Time'][i] = 'NO'
```

```python
df.head()
```

```python
df = df.drop(columns = ["Date", "Player", "Declined", "Offsetting"])
```

```python
df.loc[df['Pos'] == '', 'Pos'] = 'NP'
```

```python
df[["Quarter", "Down", "Distance", "Yards"]] = df[["Quarter", "Down", "Distance", "Yards"]].astype(int)
```

```python
df['Time'] = df['Time'].apply(lambda x: int(x.split(":")[0]) * 60 + int(x.split(":")[1]))
```

```python
df = df.rename(columns = {'Time': 'Time left'})
```

```python
df['Time left'] = df['Time left'].astype(int)
```

```python
for c in df.select_dtypes('integer'):
    
    df[c]=pd.to_numeric(df[c], downcast='integer')
```
